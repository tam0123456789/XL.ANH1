```python
import imageio.v2 as imageio
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi

img = imageio.imread("bone.png")

m1 = np.where(img > 57, 1, 0)

plt.imshow(m1, cmap="gray")

plt.axis("off")

plt.show()
```

```python
img = imageio.imread("bone.png")

m1 = np.where(img > 57, 1, 0)

dilated_m1 = ndi.binary_dilation(m1, iterations=5)

plt.imshow(dilated_m1, cmap="gray")

plt.axis("off")

plt.show()
```

```python
img = imageio.imread("bone.png")

m1 = np.where(img > 57, 1, 0)

erosion_m1 = ndi.binary_erosion(m1, iterations=5)

plt.imshow(erosion_m1, cmap="gray")

plt.axis("off")

plt.show()
```

```python
mask_bone = img >= 57
img_bone = np.where(mask_bone, img, 0)

plt.imshow(img_bone, cmap="gray")

plt.axis("off")

plt.show()
```

```python
hist = ndi.histogram(img_bone, min=0, max=255, bins=256)

plt.plot(hist)

plt.show()
```

```python
hist = ndi.histogram(img_bone, min=57, max=255, bins=255-57+1)

plt.plot(hist)

plt.show()
```

```python
mask_bone = img >= 57
mask_dilate = ndi.binary_dilation(mask_bone, iterations=5)
mask_closed = ndi.binary_closing(mask_bone, iterations=5)

fig, axes = plt.subplots(1, 3)

axes[0].imshow(mask_bone, cmap="gray")
axes[1].imshow(mask_dilate, cmap="gray")
axes[2].imshow(mask_closed, cmap="gray")

for ax in axes:
  ax.axis("off")

  plt.show()
```

```python
weights = [[.11, .11, .11],
           [.11, .12, .11],
           [.11, .11, .11]]

img_filt = ndi.convolve(img, weights)

fig, axes = plt.subplots(1, 2)

axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[1].imshow(img_filt, cmap="gray")
axes[1].set_title("Sharpened Image")

for ax in axes:
  ax.axis("off")

plt.show()
```

```python
img_filt = ndi.median_filter(img, size=10)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[1].imshow(img_filt, cmap="gray")
axes[1].set_title("Smoothed Image (with Median Filter)")

for ax in axes:
  ax.axis("off")

plt.show()
```

```python
img_filt = ndi.gaussian_filter(img, sigma=10)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[1].imshow(img_filt, cmap="gray")
axes[1].set_title("Smoothed Image (with Gaussian Filter)")

for ax in axes:
  ax.axis("off")

plt.show()
```
