```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt

img = imageio.imread("bone.png")

print("Image shape:", img.shape)
print("Minimum Pixel Value:", img.min())
print("Maximum Pixel Value:", img.max())

plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import scipy.ndimage as ndi

hist = ndi.histogram(img, min=0, max=255, bins=256)
plt.plot(hist)
plt.show()
```

```python
hist = ndi.histogram(img, min=0, max=25, bins=26)
plt.plot(hist)
plt.show()
```

```python
hist = ndi.histogram(img, min=0, max=255, bins=256)

cdf = hist.cumsum() / hist.sum()

equalized_img = cdf[img] * 255

fig, axes = plt.subplots(1, 2)

axes[0].imshow(img, cmap="gray")
axes[1].imshow(equalized_img, cmap="gray")

for ax in axes:
  ax.axis("off")

plt.show()
```

```python
hist = ndi.histogram(img, min=0, max=255, bins=256)

cdf = hist.cumsum() / hist.sum()

fig, axes = plt.subplots(2, 1)

axes[0].plot(hist, label="Histogram")
axes[1].plot(cdf, label="CDF")
plt.show()
```

```python
mask1 = img > 32
plt.imshow(mask1, cmap="gray")
plt.axis("off")
plt.show()
```

```python
mask2 = img > 64
plt.imshow(mask1, cmap="gray")
plt.axis("off")
plt.show()
```

```python
mask3 = mask1 & ~mask2
plt.imshow(mask3, cmap="gray")
plt.axis("off")
plt.show()
```
